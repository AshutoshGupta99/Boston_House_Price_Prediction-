```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error
```

```python
#generating n evenly spaced values from 0 to 2pi
n=100
x=np.linspace(0,2*np.pi,n)
sine_values = np.sin(x)

#plot sine wave
plt.plot(x,sine_values)
```

```python
#adding noise to sine wave
noise=0.5
noisy_sine_values = sine_values + np.random.uniform(-noise,noise,n)

#plot the noisy values
plt.plot(x,noisy_sine_values,color='r')
plt.plot(x,sine_values,linewidth=3)
```

```python
#Using formula to calculate MSE
error = (1/n) * np.sum(np.power(sine_values-noisy_sine_values,2))
error
```

```python
#calculating MSE using function from sklearn library
mean_squared_error(sine_values,noisy_sine_values)
```
